## K-Nearest Neighbours

### 1. Classification

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Evaluate for different K values
k_range = range(1, 21)
accuracies = []
```

```python
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
```

```python
best_k = k_range[accuracies.index(max(accuracies))]
best_acc = max(accuracies)
print(f"Best K: {best_k}, Accuracy: {best_acc:.4f}")
```

```python
plt.plot(k_range, accuracies, marker='o')
plt.title("K-NN Classification Accuracy vs K")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Accuracy")
plt.grid()

# Highlight best K
plt.scatter(best_k, best_acc, color='red', s=100, label=f"Best K = {best_k}")
plt.legend()
plt.show()
```

### 2. Regression

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Load dataset
data = fetch_california_housing()
X, y = data.data, data.target
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Evaluate different K values
k_range = range(1, 21)
mse_scores = []
```

```python
for k in k_range:
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)
```

```python

# Plot MSE vs K
plt.plot(k_range, mse_scores, marker='o', color='red')
plt.title('k-NN Regression MSE vs K')
plt.xlabel('K value')
plt.ylabel('Mean Squared Error')
plt.grid(True)
plt.show()
```

```python

# Best K evaluation
best_k = mse_scores.index(min(mse_scores)) + 1
print(f"Best K: {best_k}, Lowest MSE: {min(mse_scores):.4f}")
```
